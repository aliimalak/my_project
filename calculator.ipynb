```
def add(x,y):
  return x + y

def sub(x,y):
  return x - y

def multi(x,y):
  return x * y

def div(x,y):
  return x / y


while True:
  print("1 => ADD")
  print("2 => Sub")
  print("3 => Multi")
  print("4 => Div")
  print("5 => Exit")
  print("Choose an Option.")

  op = int(input("Enter your Option: "))

  while op > 5 or op < 1:
    print("Sorry.\nThis is not an option.")
    op = int(input("Enter your Option: "))

  num1 = float(input("Enter the 1st number: "))
  num2 = float(input("Enter the 2nd number: "))

  if op == 5:
    break

  if op == 1:
    print(f"{num1} + {num2} = {add(num1,num2)}")

  if op == 2:
    print(f"{num1} - {num2} = {sub(num1,num2)}")

  if op == 3:
    print(f"{num1} x {num2} = {multi(num1,num2)}")

  if op == 4:
    if num2 == 0:
      print("Sorry you can not divide by Zero.")
    else:
      print(f"{num1} / {num2} = {div(num1,num2)}")


















```
